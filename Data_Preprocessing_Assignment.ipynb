5) Handling Missing Data

```python
# Checking blank Values
df.isnull()
```

```python
# Checking null values category wise
df.isnull().sum()
```

```python
# Dropping categorical data rows with missing values
df.dropna(how='any', subset=['revenues', 'profits'], inplace=True)
```

```python
# Checking null values category wise
df.isnull().sum()
```

```python
# Splitting dataset into independent & dependent variable
X = df[['name','sector','rank']].values
y = df['ceowoman'].values
print(X)
```

```python
print(y)
```

```python
from sklearn.impute import SimpleImputer
print(X[:, 1:3]) #working with only numerical data
```

```python
# Handling Categorical Data
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
# [0] signifies the index of the column we are appliying the encoding on
df = pd.DataFrame(ct.fit_transform(df))
print(df)
```

```python
# Label Encoding
# Encoding the target variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
# Splitting Data for Training & Testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```
